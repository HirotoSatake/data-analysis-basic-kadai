```python
import matplotlib.pyplot as plt
import numpy as np

# サブプロットの作成
fig, axes = plt.subplots(2, 3, figsize=(12, 8), tight_layout=True)

# サブプロット（0, 0）
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)
axes[0, 0].set_title("Line Plot")

# サブプロット(0, 1)
y1 = [1, 2, -5, 2]
axes[0, 1].plot(y1)
axes[0, 1].hist(y1, bins=50)  # histはaxesを通して描画
axes[0, 1].set_title("Histogram")

# サブプロット(0, 2)
y2 = [1, 2, -5, 2]
axes[0, 2].plot(y2)
axes[0, 2].boxplot(y2)  # boxplotもaxesを通して描画
axes[0, 2].set_title("Boxplot")

# サブプロット(1, 0)
scatter_x = [1, 2, -5, 2]
scatter_y = [4, 5, 10, 2]
axes[1, 0].scatter(scatter_x, scatter_y)
axes[1, 0].set_title("Scatter Plot")

# サブプロット(1, 1)
y4 = np.array([1, 2, 5, 3])
labels = ["dog", "bird", "cat", "elephant"]
axes[1, 1].pie(y4, labels=labels, autopct='%1.1f%%')
axes[1, 1].set_title("Favorite Animals")

# サブプロット(1, 2)
x_vals = np.array([1, 2, 3, 4, 5])
y_vals = np.array([100, 123, 231, 234, 110])
labels = ["beef", "pork", "shark", "chicken", "salmon"]
axes[1, 2].bar(x_vals, y_vals, tick_label=labels)
axes[1, 2].set_title("Favorite Food")
axes[1, 2].legend(labels=labels)  # 凡例を追加

# 表示
plt.show()
```
