```python
import matplotlib.pyplot as plt
import numpy as np

#
fig, axes = plt.subplots(2,3, figsize=(10,8),tight_layout=True)

#0,0
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

#(0,1)
y1 = [1,2,-5,2,3,6,5,3,2,4,6]
axes[0,1].hist(y1, bins=50)

#(0,2)
y2 = [1, 2, -5, 2,5,7,8,4]
axes[0,2].boxplot(y2)

#(1,0)
scatter_x = [1, 2, -5, 2]
scatter_y = [4,5,10,2]
axes[1,0].scatter(scatter_x, scatter_y)
axes[1,0].set_title("scatter plot")


#(1,1)
y4=np.array([1,2,5,3])
label = ["dog","bird","cat","elephant"]
axes[1,1].set_title("favorite animal")
axes[1,1].pie(y4, labels=label,autopct='%1.1f%%')

#(1,2)
x = np.array([1,2,3,4,5])
y = np.array([100,123,231,234,110])
label = ["beef","pork","shark","chiken","salmon"]
axes[1,2].bar(x,y,tick_label=label)
axes[1,2].set_title("favorite food")



plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np


fig, axes = plt.subplots(nrows=2, ncols=3, tight_layout=True)

# 0,0
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)


#sin0,1
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)
axes[0,1].plot(x, y, linewidth=2.0)
axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
              ylim=(-1, 5), yticks=np.arange(-1, 6))


# 0,2
data = np.random.normal(15,5,2000)
axes[0, 2].hist(data, bins=30)


# 1,0
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 60)
y = 4 + np.random.normal(0, 2, len(x))
axes[1, 0].scatter(x, y)



# 
plt.show()
```
